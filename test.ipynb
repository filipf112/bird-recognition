```python
import xenocanto
import requests
```

```python
def search_xeno_canto(query):
    base_url = "https://www.xeno-canto.org/api/2/recordings"
    params = {"query": query}
    response = requests.get(base_url, params=params)
    
    if response.status_code == 200:
        data = response.json()
        return data
    else:
        return None

def main():
    query = input("Enter your bird query: ")
    results = search_xeno_canto(query)
    
    if results:
        print("Results:")
        for idx, recording in enumerate(results["recordings"]):
            print(f"{idx+1}. Title: {recording['en']}, Country: {recording['cnt']}, Recording ID: {recording['id']}")
    else:
        print("No results found.")

if __name__ == "__main__":
    main()
```

```python
import requests

def download_recording(recording_id):
    base_url = f"https://www.xeno-canto.org/{recording_id}/download"
    response = requests.get(base_url)
    
    if response.status_code == 200:
        with open(f"{recording_id}.mp3", "wb") as f:
            f.write(response.content)
        print("Recording downloaded successfully.")
    else:
        print("Failed to download recording.")

def main():
    recording_id = input("Enter the recording ID: ")
    download_recording(recording_id)

if __name__ == "__main__":
    main()
```
